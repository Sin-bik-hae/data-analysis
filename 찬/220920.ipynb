# PCA

```python
from sklearn.preprocessing import StandardScaler  # 표준화 패키지 라이브러리 
x = df_b.iloc[:,:-2].values # 독립변인들의 value값만 추출
y = df_p['P3'].values # 종속변인 추출

x = StandardScaler().fit_transform(x) # x객체에 x를 표준화한 데이터를 저장

features = df_b.columns[:-2]
pd.DataFrame(x, columns=features).head()
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2) # 주성분을 몇개로 할지 결정
printcipalComponents = pca.fit_transform(df_b.values)
principalDf = pd.DataFrame(data=printcipalComponents)
principalDf
# 주성분으로 이루어진 데이터 프레임 구성
```

```python
# pca.explained_variance_ratio_
y
```

```python
import matplotlib.pyplot as plt

fig = plt.figure(figsize = (8, 8))
ax = fig.add_subplot(1, 1, 1)
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('2 component PCA', fontsize=20)

targets = [1, 0]
colors = ['r', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = y == target
    ax.scatter(principalDf.loc[indicesToKeep, 0]
               , principalDf.loc[indicesToKeep, 1]
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
```
